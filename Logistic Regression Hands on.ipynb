```python
#checking outlier

for i in df.columns:
    if df[i].dtype!='object':
        sns.boxplot(df[i])
        plt.xlabel(i)
        plt.show()
```

```python
out_col=['age','duration','campaign','previous']
```

```python
for i in out_col:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    
    df=df[(df[i]>=lower) & (df[i]<=upper)]
```
